```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
img1= cv2.imread('river1.jpg')
gray1= cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
sift1 = cv2.xfeatures2d.SIFT_create()
kp1, des1 = sift1.detectAndCompute(gray1,None)
img1=cv2.drawKeypoints(gray1,kp1,img1,flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(img1)
plt.show()
cv2.imwrite('sift_keypoints.jpg',img1)

img2 = cv2.imread('river2.jpg')
gray2= cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
sift2 = cv2.xfeatures2d.SIFT_create()
kp2, des2 = sift2.detectAndCompute(gray2,None)
img2=cv2.drawKeypoints(gray2,kp2,img2,flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(img2)
plt.show()
cv2.imwrite('sift_keypoints2.jpg',img2)
```

```python
print(des1.shape)
print(des2.shape)
```

```python
maxscale=0
index1=0
for i in range (des1.shape[0]):
    if np.linalg.norm(des1[i]) > maxscale:
        maxscale=np.linalg.norm(des1[i])
        index1=i
print(maxscale)
print(index1)
```

```python
mindist=10000
index2=0
for i in range (des2.shape[0]):
    if np.linalg.norm(des1[index1]-des2[i]) <mindist:
        mindist=np.linalg.norm(des1[index1]-des2[i])
        index2=i
print(mindist)
print(index2)
```

```python
bf = cv2.BFMatcher(cv2.NORM_L2,crossCheck=False)
D1=[]
D1.append(des1[index1]);
D2=[]
D2.append(des2[index2]);
D1=np.array(D1)
D2=np.array(D2)
matches = bf.match(D1,D2)
print(matches)
img3 = cv2.drawMatches(img1,kp1[index1:index1+1],img2,kp2[index2:index2+1],matches,None, flags=2)
plt.imshow(img3),plt.show()
cv2.imwrite('matches.jpg',img3)
```

```python
#second method:
bf = cv2.BFMatcher(cv2.NORM_L2,crossCheck=False)
matches = bf.match(des1,des2)
img3 = cv2.drawMatches(img1,kp1,img2,kp2,matches[index1:index1+1],None, flags=2)
img3 = cv2.drawMatches(img1,kp1,img2,kp2,matches[index1:index1+1],None, flags=2)
plt.imshow(img3),plt.show()
cv2.imwrite('matches2.jpg',img3)
```
